# Part 1

```python
# read the input file
with open("input.txt") as f:
    data = f.read()

data="""162,817,812
57,618,57
906,360,560
592,479,940
352,342,300
466,668,158
542,29,236
431,825,988
739,650,466
52,470,668
216,146,977
819,987,18
117,168,530
805,96,715
346,949,466
970,615,88
941,993,340
862,61,35
984,92,344
425,690,689"""
data = data.splitlines()

for index, line in enumerate(data):
    print(f"line {index}: {line}")
```

```python
# read in each line as one node containing a x y and z coordinate with possible edges to other nodes
nodes = []
for line in data:
    x, y, z = map(int, line.split(","))
    nodes.append((x, y, z))

print(nodes)
```

```python
# funktion to get the straight-line (Euclidean) distance between two nodes
def cosine_distance(node1, node2):
    from math import sqrt
    return sqrt(sum((a - b) ** 2 for a, b in zip(node1, node2)))
```

```python
# calculate all possible cosine distances between nodes
distances = {}
for i in range(len(nodes)):
    for j in range(i + 1, len(nodes)):
        dist = cosine_distance(nodes[i], nodes[j])
        distances[(i, j)] = dist

print(distances)
```

```python
# append a edge between the nodes that are the closest to each other
edges = []
number_of_edges = 1000
sorted_distances = sorted(distances.items(), key=lambda x: x[1])
for (i, j), dist in sorted_distances[:number_of_edges]:
    edges.append((i, j, dist))

print(edges)
```

```python
# count how many clusters are formed by the edges. Nodes are in the same cluster if they are connected by an edge
clusters = []
for i in range(len(nodes)):
    clusters.append({i})

for i, j, dist in edges:
    # find the clusters that contain i and j
    cluster_i = None
    cluster_j = None
    for cluster in clusters:
        if i in cluster:
            cluster_i = cluster
        if j in cluster:
            cluster_j = cluster
    # if they are different clusters, merge them
    if cluster_i != cluster_j:
        cluster_i.update(cluster_j)
        clusters.remove(cluster_j)

print(f"Number of clusters: {len(clusters)}")

for cluster in clusters:
    print(cluster)
```

```python
# get the length of the three largest clusters
largest_clusters = sorted(clusters, key=lambda x: len(x), reverse=True)[:3]
lengths = [len(cluster) for cluster in largest_clusters]
print(f"Lengths of the three largest clusters: {lengths}")
```

```python
# multiply the lengths together to get the final result
result = 1
for length in lengths:
    result *= length

print(f"Final result: {result}")
```

```python
# read in each line as one node containing a x y and z coordinate with possible edges to other nodes
nodes = []
for line in data:
    x, y, z = map(int, line.split(","))
    nodes.append((x, y, z))

print(nodes)
```

```python
# funktion to get the straight-line (Euclidean) distance between two nodes
def cosine_distance(node1, node2):
    from math import sqrt
    return sqrt(sum((a - b) ** 2 for a, b in zip(node1, node2)))
```

```python
# calculate all possible cosine distances between nodes
distances = {}
for i in range(len(nodes)):
    for j in range(i + 1, len(nodes)):
        dist = cosine_distance(nodes[i], nodes[j])
        distances[(i, j)] = dist

print(distances)
```

```python
# append a edge between the nodes that are the closest to each other
edges = []
sorted_distances = sorted(distances.items(), key=lambda x: x[1])
for (i, j), dist in sorted_distances:
    edges.append((i, j, dist))

print(edges)
```

```python
# count how many clusters are formed by the edges. Nodes are in the same cluster if they are connected by an edge
clusters = [{i} for i in range(len(nodes))]

for step, (i, j, dist) in enumerate(edges, start=1):
    print(f"Step {step}: processing edge ({i}, {j}) at distance {dist}")
    # find the clusters that contain i and j
    cluster_i = cluster_j = None
    for cluster in clusters:
        if i in cluster:
            cluster_i = cluster
        if j in cluster:
            cluster_j = cluster
    # if they are different clusters, merge them
    if cluster_i is not cluster_j:
        cluster_i.update(cluster_j)
        clusters.remove(cluster_j)
    # print current clusters after this iteration
    print(f"Clusters after step {step}:\n")
    for c in clusters:
        print(c)
    if len(clusters) == 1:
        print("All nodes are now in a single cluster.")
        print(f"\n\nFinal connection:\n")
        print(f"Node {i} connected to Node {j} at distance {dist}")
        break

print(f"Number of clusters: {len(clusters)}")
for cluster in clusters:
    print(cluster)
```
