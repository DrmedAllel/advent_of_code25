# Part 1

```python
# Read in all the data from the input file
with open('input.txt') as f:
    input_data = f.read().strip()

# input_data = """3-5
# 10-14
# 16-20
# 12-18

# 1
# 5
# 8
# 11
# 17
# 32
# """

print(input_data)
```

```python
# Split the input data after \n in two parts
parts = input_data.split('\n\n')
ranges_data = parts[0].split('\n')
ingredient_data = parts[1].split('\n')

print("Ranges Data:")
print(ranges_data)
print("Ingredient Data:")
print(ingredient_data)
```

```python
borders = []
for range_str in ranges_data:
    border = range_str.split('-')
    borders.append(border)

print("Borders:")
print(borders)
```

```python
fresh_ingredients = []

for ingredient in ingredient_data:
    if ingredient == '':
        continue  # skip empty strings
    for border in borders:
        if int(border[0]) <= int(ingredient) <= int(border[1]):
            fresh_ingredients.append(ingredient)
            break  # ingredient found in a range, no need to check further

print("Number of fresh ingredients:")
print(len(fresh_ingredients))
```

# Part 2

```python
# combine overlapping ranges
merged_borders = []
for border in sorted(borders, key=lambda x: int(x[0])):
    if not merged_borders or int(merged_borders[-1][1]) < int(border[0]):
        merged_borders.append(border)
    else:
        merged_borders[-1][1] = str(max(int(merged_borders[-1][1]), int(border[1])))

print("Merged Borders:")
print(merged_borders)
```

```python
# for every range calculate the number of ingredients that could fall within that range
number_of_ingredients = 0
for border in merged_borders:
    number_of_ingredients += int(border[1]) - int(border[0]) + 1
    
print("Total number of possible ingredients:")
print(number_of_ingredients)
```
