# Part 1

```python
# read in the input file
with open("input.txt", "r", encoding="utf-8") as fh:
    raw_input = fh.read()



print(raw_input)

```

```python
# make a list of lines
input = raw_input.splitlines()

print(input)
```

```python
count_zero = 0
dial_position = 50

for line in input:
    if "L" in line:
        line = int(line.replace("L", "-"))
    elif "R" in line:
        line = int(line.replace("R", "+"))

    dial_helper = dial_position + line
    dial_position = dial_helper % 100
    if dial_position == 0:
        count_zero += 1

print("Number of times the dial lands on 0:", count_zero)


    
        
```

# Part 2

```python
count_zero = 0
dial_position = 50

input = raw_input.splitlines()

for line in input:
    if "L" in line:
        move = int(line.replace("L", "-"))
    elif "R" in line:
        move = int(line.replace("R", "+"))

    for step in range(abs(move)):
        if move < 0:
            dial_position -= 1
        else:
            dial_position += 1

        dial_position = dial_position % 100
        if dial_position == 0:
            count_zero += 1

print("Number of times the dial lands on 0 (part 2):", count_zero)
```
