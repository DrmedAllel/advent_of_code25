# Part 1


```python
# load the input file
with open("input.txt", "r") as f:
    raw_input = f.read()

ID_Pairs = raw_input.split(",")

print(ID_Pairs)
```

```python
# Seperate the ID pairs into two independent numbers

IDs = []

for pair in ID_Pairs:
    id1, id2 = pair.split("-")
    IDs.append((int(id1), int(id2)))

print(IDs)
```

```python
false_ids = []

for id_pair in IDs:
    id1, id2 = id_pair

    # Create a new list with all numbers between id1 and id2 (inclusive)
    id_range = list(range(id1, id2 + 1)) 
    for number in id_range:
        # if the number of digits is odd continue
        if len(str(number)) % 2 != 0:
            continue
        #split the number in half and save as two separate numbers
        str_number = str(number)
        half = len(str_number) // 2
        first_half = str_number[:half]
        second_half = str_number[half:]

        if first_half == second_half:
            false_ids.append(int(number))
            continue

print("Sum of all false IDs:", sum(false_ids))
```

# Part 2



```python
false_ids = []

# Add the specified pairs and their invalid IDs to the array
# IDs = [(11,22), (95,115), (998,1012), (1188511880,1188511890), (222220,222224), (1698522,1698528), (446443,446449), (38593856,38593862), (565653,565659), (824824821,824824827), (2121212118,2121212124)]

for id_pair in IDs:
    id1, id2 = id_pair

    # Create a new list with all numbers between id1 and id2 (inclusive)
    id_range = list(range(id1, id2 + 1)) 
    for number in id_range:
        # get the length of the number
        length = len(str(number))

        num_of_parts = 2
        while num_of_parts <= length:
            # split the number into num_of_parts parts
            str_number = str(number)
            part_length = length // num_of_parts
            # if all parts are equal, add to false_ids
            parts = [str_number[i:i + part_length] for i in range(0, length, part_length)]
            if all(part == parts[0] for part in parts):
                false_ids.append(int(number))
                break
            num_of_parts += 1  

print(false_ids)
print("Sum of all false IDs:", sum(false_ids))
```
