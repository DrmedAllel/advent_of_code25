# Part 1

```python
# Load the input data
with open('input.txt') as f:
    raw_data = f.read()

raw_data = """123 328  51 64 
 45 64  387 23 
  6 98  215 314
*   +   *   +  
"""


print("Raw Data:")
print(raw_data)
```

```python
lines = raw_data.split('\n')
print("Lines:")
for line in lines:
    print(line + '\n')
```

```python
lines = [line for line in lines if line.strip() != '']
row_index = 0
matrix = []
for line in lines:
    elements = line.split(" ")
    elements = [el for el in elements if el != '']
    matrix.append(elements)
print("Combined matrix:")
for idx, row in enumerate(matrix):
    print(f"Row {idx}: {row}")
    
```

```python
# select the problems by saving all elements from one column in a list
problems = []
for col_idx in range(len(matrix[0])):
    column_elements = []
    for row in matrix:
        column_elements.append(row[col_idx])
    problems.append(column_elements)
print("Problems:")
print(problems)
```

```python
solutions = []
for problem in problems:
    if '*' in problem:
        result = 1
        for el in problem:
            if el != '*':
                result *= int(el)
    elif '+' in problem:
        result = 0
        for el in problem:
            if el != '+':
                result += int(el)
    solutions.append(result)

print("Sum of solutions:")
print(sum(solutions))
```

# Part 2

```python
# make sure that all numbers in the problems have the same length by padding with leading zeros
max_length = 0
for problem in problems:
    for el in problem:
        if el not in ['*', '+']:
            max_length = max(max_length, len(el))
for i in range(len(problems)):
    for j in range(len(problems[i])):
        if problems[i][j] not in ['*', '+']:
            problems[i][j] = problems[i][j].zfill(max_length)
print("Padded Problems:")
print(problems)
```

```python
# build the new numbers by going character for character from top to bottom
for problem in problems:
    numbers = []
    for number in problem:
        if number not in ['*', '+']:
            numbers.append(number)
    amount_of_numbers = len(numbers)
    length = len(numbers[0])
    # put all numbers together in one string
    combined_number = ''
    for i in range(length):
        for number in numbers:
            combined_number += number[i]
    print("Combined Number:")
    print(combined_number)

    # get the numbers apart again
    new_numbers = []
    i = amount_of_numbers
    for i in range(amount_of_numbers):
        new_number = combined_number[i*length:(i+1)*length]
        reversed_number = new_number[::-1]
        new_numbers.append(reversed_number)
    print("New Numbers:")
    print(new_numbers)
```
