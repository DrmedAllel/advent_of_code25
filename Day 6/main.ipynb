# Part 1

```python
# Load the input data
with open('input.txt') as f:
    raw_data = f.read()

# raw_data = """123 328  51 64 
#  45 64  387 23 
#   6 98  215 314
# *   +   *   +  
# """


print("Raw Data:")
print(raw_data)
```

```python
lines = raw_data.split('\n')
print("Lines:")
for line in lines:
    print(line + '\n')
```

```python
lines = [line for line in lines if line.strip() != '']
row_index = 0
matrix = []
for line in lines:
    elements = line.split(" ")
    elements = [el for el in elements if el != '']
    matrix.append(elements)
print("Combined matrix:")
for idx, row in enumerate(matrix):
    print(f"Row {idx}: {row}")
    
```

```python
# select the problems by saving all elements from one column in a list
problems = []
for col_idx in range(len(matrix[0])):
    column_elements = []
    for row in matrix:
        column_elements.append(row[col_idx])
    problems.append(column_elements)
print("Problems:")
print(problems)
```

```python
solutions = []
for problem in problems:
    if '*' in problem:
        result = 1
        for el in problem:
            if el != '*':
                result *= int(el)
    elif '+' in problem:
        result = 0
        for el in problem:
            if el != '+':
                result += int(el)
    solutions.append(result)

print("Sum of solutions:")
print(sum(solutions))
```

# Part 2

```python
lines = raw_data.split('\n')
print("Lines:")
for line in lines:
    print(line + '\n')
```

```python
lines = [line for line in lines if line.strip() != '']
row_index = 0
matrix = []
for line in lines:
    elements = line.split(" ")
    elements = [el for el in elements if el != '']
    matrix.append(elements)
print("Combined matrix:")
for idx, row in enumerate(matrix):
    print(f"Row {idx}: {row}")
```

```python
# for each column, find the length of the longest element
max_lengths = []
for col_idx in range(len(matrix[0])):
    max_length = 0
    for row in matrix:
        el_length = len(row[col_idx])
        if el_length > max_length:
            max_length = el_length
    max_lengths.append(max_length)
print("Max lengths per column:")
print(max_lengths)
```

```python
lines = [line for line in lines if line.strip() != '']
row_index = 0
rows = []
for line in lines:
    elements = line.split(" ")
    elements = [el if el != '' else '0' for el in elements]
    # save all elements from one row in one string
    string = ''
    for character in elements:
        string += character
    rows.append(string)
print("Rows as strings:")
for row in rows:
    print(row)
```

```python
numbers = rows[:-1]  # all rows except the last one
operations = rows[-1]  # the last row 
problems = []

for max_length in max_lengths:
    # cut the first max_length characters from operations
    action = operations[:max_length]
    operations = operations[max_length:]
    

    # go through each row and save the first max_length characters
    elements = []
    for idx, row in enumerate(numbers):
        element = row[:max_length]
        elements.append(element)

        # update the row by removing the first max_length characters
        numbers[idx] = row[max_length:]

    # reverse the elements to maintain original order
    elements = elements[::-1]
    problems.append((action, elements))

print("Problems:")
for problem in problems:
    print(problem)
        
```

```python
matrices_with_action = []
for problem in problems:
    action, elements = problem
    matrix = []
    for el in elements:
        matrix.append(el)
    matrices_with_action.append((action, matrix))

for action, matrix in matrices_with_action:
    print(f"Action: {action}")
    print("Matrix:")
    for row in matrix:
        print(row)
```

```python
def reverse_int(n: int) -> int:
    s = str(n)
    sign = -1 if s[0] == '-' else 1
    rev = s.lstrip('-')[::-1]
    return sign * int(rev)
```

```python
# from each matrix extact one column as a new number of the problem
problems = []
for action, matrix in matrices_with_action:
    column_elements = []
    for i in range(len(matrix[0])):
        column_characters = []
        for row in matrix:
            # remove leading zeros from each row element
            number = row[i].lstrip('0')
            column_characters.append(number)
        column_elements.append(reverse_int(''.join(column_characters)))
    problems.append((action, column_elements))

print("Final Problems:")
for action, elements in problems:
    print(f"Action: {action}, Elements: {elements}")
```

```python
final_solutions = []
for action, elements in problems:
    if '*' in action:
        result = 1
        for el in elements:
            result *= el
    elif '+' in action:
        result = sum(elements)
    final_solutions.append(result)

print("Final Solutions:")
for sol in final_solutions:
    print(sol)
```

```python
sum_of_final_solutions = sum(final_solutions)
print("Sum of Final Solutions:")
print(sum_of_final_solutions)
```
