# Part 1


```python
# Read in the input file
with open('input.txt', 'r') as file:
    input_data = file.readlines()
    input_data = [line.strip() for line in input_data]

print(input_data)
```

```python
max_combinations = []

#input_data = ['987654321111111','811111111111119','234234234234278','818181911112111']

for line in input_data:
    # get the maximum number from the line
    line_minus_one = line[:-1]
    first_max_number = max(int(character) for character in line_minus_one)  
    
    # get the part of the line after the maximum number
    first_max_index = line_minus_one.index(str(first_max_number))
    line_after_max = line[first_max_index + 1:]

    # pick the maximum number from the second part
    if line_after_max:
        second_max_number = max(int(character) for character in line_after_max)

    max_combination = int(f"{first_max_number}{second_max_number}")

    max_combinations.append(max_combination)


print("Max combinations:", max_combinations)
print("Sum of max combinations:", sum(max_combinations))


        


```

# Part 2


```python
max_combinations = []

#input_data = ['987654321111111','811111111111119','234234234234278','818181911112111']

for line in input_data:
    digits = []
    digits_per_battery = 12
    
    while digits_per_battery > 0:
        # Cut the last digits off the line
        line_minus_length_battery = line[:-(digits_per_battery - 1)] if digits_per_battery > 1 else line
        max_digit = max(int(character) for character in line_minus_length_battery)
        index_max_digit = line_minus_length_battery.index(str(max_digit))
        digits.append(max_digit)
        line = line[index_max_digit + 1:]
        digits_per_battery -= 1
    
    max_combo = ""
    for digit in digits:
        max_combo += str(digit)
    
    max_combinations.append(int(max_combo))
print("Max combinations:", max_combinations)
print("Sum of max combinations:", sum(max_combinations))
```
